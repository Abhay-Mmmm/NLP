**Spacy is Object Oriented**

Provides most efficient NLP algos for a given task.

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("Dr. Strange love pav bhaji. Hulk loves momos of Assam.")

for sentence in doc.sents:
    print(sentence)
```

```python
for sentence in doc.sents:
  for word in sentence:
    print(word)
```
