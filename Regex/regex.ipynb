```python
import re
```

```python
chat1 = 'codebasics: you ask lot of questions  12345678912, abc@xyz.com, 9998881234'
chat2 = 'codebasics: here it is :   (123)-567-8912, abc_08@xyz.com'
chat3 = 'codebasics: yes, phone: 12345678912 email: abc0@xyz.com'
```

```python
pattern = '\d{10}'

matches = re.findall(pattern, chat1)
matches
```

```python
pattern = '\d{10}|\(\d{3}\)-\d{3}-\d{4}' # | symbolizes OR

matches = re.findall(pattern, chat3)
matches
```

```python
pattern = '[a-z0-9A-Z_]*@[a-z0-9A-Z]*\.[a-zA-Z]*'

matches = re.findall(pattern, chat1)
email = matches[0]
email
```
