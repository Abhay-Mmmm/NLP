```python
import spacy
```

```python
nlp = spacy.blank("en")

doc = nlp("Dr. Strange loves pav bhaji of mumbai as it costs only 2$.")

for token in doc:
    print(token)
```

```python
with open("student.txt") as f:
    text = f.readlines()
text
```

```python
text = ' '.join(text)
text
```

```python
doc = nlp(text)
email = []
for token in doc:
  if token.like_email:
    email.append(token)
email
```

```python
doc = nlp("gimme double cheese extra large healthy pizza")

tokens = [token.text for token in doc]
tokens
```

```python
from spacy.symbols import ORTH

nlp.tokenizer.add_special_case("gimme",[
    {ORTH:"gim"},
    {ORTH:"me"}
])

doc = nlp("gimme double cheese extra large healthy pizza")
tokens = [token.text for token in doc]
tokens
```

```python
nlp.add_pipe('sentencizer')
```

```python
doc = nlp("Dr. Strange loves pav bhaji of mumbai as it costs only $2. Tony Stark hates the people here.")

for sentence in doc.sents:
  print(sentence)
  print()
```
